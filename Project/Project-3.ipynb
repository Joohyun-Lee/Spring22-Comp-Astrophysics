```python
def _MagneticFieldStrength(field, data):
    return (
        (ad[('all','MagneticField')][:,0]**2+ad[('all','MagneticField')][:,1]**2+ad[('all','MagneticField')][:,2]**2)**0.5
    )

ds.add_field(
    name=("all", "MagneticFieldStrength"),
    function=_MagneticFieldStrength,
    sampling_type="local",
)
```
